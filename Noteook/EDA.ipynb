# "data2" - group by users&date

**"data2"** label each customer as new, retained,resurrected every time they purchase.
It contains following information of each each customer's purchase information in each day: 
* the number of total session
* the number of total product, 
* the number of total product brand, 
* the number of total product category 
* mean price for each product
* number of days since last purchase time
* label

```python
#group by user_id and star_days
data2 = data1.groupby(['user_id', 'date']).agg({'user_session':'count','event_type':'count', 
                                                     'product_id':'count', 'category_id':'count', 'brand':'count', 'price':'count', 'price':'mean'})
#reset index
data2.reset_index(inplace=True)

#change to date type
data2["date"] = data2["date"].astype("datetime64")

# rename column
data2 = data2.rename(columns = {'user_session':'count_user_session', 
                                'event_type':'count_event_type', 
                                'product_id':'count_product_id', 
                                'category_id':'count_category_id', 
                                'brand':'count_brand', 
                                'price':'mean_price'})

data2.head()
```

```python
# Calculate the number of days for each customer since they purchased on this website last time.
data2["difference"] = data2.groupby(["user_id"])["date"].transform(lambda x:x - x.shift(1))

data2["difference"] = data2["difference"].dt.days
# If it's their first purchase, they fill with -1/
data2["difference"] = data2["difference"].fillna(-1)
```

```python
# label each customer with new, resurrected, and retained
data2["label"] = ""
data2.label[data2["difference"] == -1] = "new"
data2.label[data2["difference"] > 7] = "resurrected" #churn
data2.label[(data2["difference"] <= 7) & (data2["difference"] >= 0)] = "retained"
```

## data3 - group by category

For each category, calculate the conversion rate, count the number of brand, and get the brand name which has the highest & lowest four conversion rates.

**Conversion rate**:
- View-Cart: ratio of number of products added to cart from viewed
- Cart-Purchase: ratio of number of products purchased from added to cart
- View-Purchase: ratio of number of products purchased from viewed
- Cart-Remove: ratio of number of products removed from added to cart

```python
# function for calculating percentage
def turn_percentage(x):
    return '%.2f%%' % (x * 100);
```

```python
brand_num = data1.groupby(['category_id']).agg({'brand':"count"})
brand_num.reset_index(inplace=True)
```

```python
brand_num
```

```python
data3 = data1.groupby(['category_id', 'event_type']).agg({'event_type':'count'})
brand_num = data1.groupby(['category_id']).agg({'brand':"count"})


data3 = data3.unstack()
data3.reset_index(inplace=True)
data3.columns = ["category_id","cart","purchase","remove_from_cart","view"]

brand_num.reset_index(inplace=True)

data3= data3.fillna(0)

data3["View-Cart"] = (data3["cart"]/data3["view"]).apply(turn_percentage)
data3["Cart-Purchase"] = (data3["purchase"]/data3["cart"]).apply(turn_percentage)
data3["View-Purchase"] = (data3["purchase"]/data3["view"]).apply(turn_percentage)
data3["Cart-Remove"] = (data3["remove_from_cart"]/data3["view"]).apply(turn_percentage)

data3 = pd.merge(data3, brand_num, how="inner", on = "category_id")

data3
```

## data4 -- group by brand

Calculate the conversion rate each brand

**Conversion rate**:
- View-Cart: ratio of number of products added to cart from viewed
- Cart-Purchase: ratio of number of products purchased from added to cart
- View-Purchase: ratio of number of products purchased from viewed
- Cart-Remove: ratio of number of products removed from added to cart

```python
data4 = data1.groupby(['category_id', "brand", 'event_type']).agg({'event_type':'count'})
data4 = data4.unstack()
data4.reset_index(inplace=True)
data4.columns = ["category_id","brand", "cart","purchase","remove_from_cart","view"]

data4= data4.fillna(0)

data4["View-Cart"] = (data4["cart"]/data4["view"]).apply(turn_percentage)
data4["Cart-Purchase"] = (data4["purchase"]/data4["cart"]).apply(turn_percentage)
data4["View-Purchase"] = (data4["purchase"]/data4["view"]).apply(turn_percentage)
data4["Cart-Remove"] = (data4["remove_from_cart"]/data4["cart"]).apply(turn_percentage)

data4
```

### Get the brand name which has the highest & lowest four conversion rates

```python
def func_max(x, y):
    return x[x[y] == x[y].max()]["brand"]

def func_min(x, y):
    return x[x[y] == x[y].min()]["brand"]
```

```python
VC_max = pd.DataFrame(data4.groupby(["category_id"]).apply(func_max, "View-Cart"))
VC_max.reset_index(inplace=True)
VC_max = VC_max.drop(["level_1"], axis = 1)
VC_max.columns =["category_id", "VC_max"]
data3 = pd.merge(data3, VC_max, how='outer', on='category_id')

VC_min = pd.DataFrame(data4.groupby(["category_id"]).apply(func_min, "View-Cart"))
VC_min.reset_index(inplace=True)
VC_min = VC_min.drop(["level_1"], axis = 1)
VC_min.columns =["category_id", "VC_min"]
data3 = pd.merge(data3, VC_min, how='outer', on='category_id')

CP_max = pd.DataFrame(data4.groupby(["category_id"]).apply(func_max, "Cart-Purchase"))
CP_max.reset_index(inplace=True)
CP_max = CP_max.drop(["level_1"], axis = 1)
CP_max.columns =["category_id", "CP_max"]
data3 = pd.merge(data3, CP_max, how='outer', on='category_id')

CP_min = pd.DataFrame(data4.groupby(["category_id"]).apply(func_min, "Cart-Purchase"))
CP_min.reset_index(inplace=True)
CP_min = CP_min.drop(["level_1"], axis = 1)
CP_min.columns =["category_id", "CP_min"]
data3 = pd.merge(data3, CP_min, how='outer', on='category_id')

VP_max = pd.DataFrame(data4.groupby(["category_id"]).apply(func_max, "View-Purchase"))
VP_max.reset_index(inplace=True)
VP_max = VP_max.drop(["level_1"], axis = 1)
VP_max.columns =["category_id", "VP_max"]
data3 = pd.merge(data3, VP_max, how='outer', on='category_id')

VP_min = pd.DataFrame(data4.groupby(["category_id"]).apply(func_min, "View-Purchase"))
VP_min.reset_index(inplace=True)
VP_min = VP_min.drop(["level_1"], axis = 1)
VP_min.columns =["category_id", "VP_min"]
data3 = pd.merge(data3, VP_min, how='outer', on='category_id')

CR_max = pd.DataFrame(data4.groupby(["category_id"]).apply(func_max, "Cart-Remove"))
CR_max.reset_index(inplace=True)
CR_max = CR_max.drop(["level_1"], axis = 1)
CR_max.columns =["category_id", "CR_max"]
data3 = pd.merge(data3, CR_max, how='outer', on='category_id')

CR_min = pd.DataFrame(data4.groupby(["category_id"]).apply(func_min, "Cart-Remove"))
CR_min.reset_index(inplace=True)
CR_min = CR_min.drop(["level_1"], axis = 1)
CR_min.columns =["category_id", "CR_min"]
data3 = pd.merge(data3, CR_min, how='outer', on='category_id')
```
